### Step 2. 리뷰에서 특수문자 제거

```python
import re
def clean_str(string):
    string = re.sub(r"[^가-힣A-Za-z0-9(),!?\'\`]", " ", string)
    string = re.sub(r"\'s", " \'s", string)
    string = re.sub(r"\'ve", " \'ve", string)
    string = re.sub(r"n\'t", " n\'t", string)
    string = re.sub(r"\'re", " \'re", string)
    string = re.sub(r"\'d", " \'d", string)
    string = re.sub(r"\'ll", " \'ll", string)
    string = re.sub(r",", " , ", string)
    string = re.sub(r"!", " ! ", string)
    string = re.sub(r"\(", " \( ", string)
    string = re.sub(r"\)", " \) ", string)
    string = re.sub(r"\?", " \? ", string)
    string = re.sub(r"\s{2,}", " ", string)
    string = re.sub(r"\'{2,}", "\'", string)
    string = re.sub(r"\'", "", string)
    return string.lower()


#EMOJI = re.compile('[\U00010000-\U0010ffff]', flags=re.UNICODE)
#TEXT_refine = EMOJI.sub(r'', text)
```

# [ 감성분석 : 긍정 / 부정 리뷰 판별하기 ]

### Step 1. 데이터 전처리 (1-3점 : 부정 / 4-5점 : 긍정)

```python
data['po_ne'] = data['score'].apply(lambda x : 0 if x <= 3 else 1)
```

```python
data
```

```python
# 4~5점(긍정) : 147,652 / 1~3점(부정) : 63,382

data['po_ne'].value_counts()
```

### Step 2. Train, Test 데이터 나누기

```python
# train 70,000개 / test 30,000개

train_df = data[:70000]
test_df = data[70000:100000]
```

### Step 3. 리뷰를 형태소 단위로 나누고 학습

```python
from konlpy.tag import Okt

Okt = Okt()

def okt_tokenizer(text):
    tokens_ko = Okt.morphs(text)
    return tokens_ko
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vect = TfidfVectorizer(tokenizer=okt_tokenizer,
                            ngram_range=(1, 2),
                            min_df = 3,
                            max_df = 0.9)

tfidf_vect.fit(train_df['review'])
tfidf_matrix_train = tfidf_vect.transform(train_df['review'])
```

```python
from lightgbm import LGBMClassifier

lgbm_clf = LGBMClassifier(n_estimators = 400)
lgbm_clf.fit(tfidf_matrix_train, train_df['po_ne'])
```

### Step 4. 테스트 -> Accuracy 측정 : 84.7%

```python
from sklearn.metrics import accuracy_score

tfidf_matrix_test = tfidf_vect.transform(test_df['review'])
preds = lgbm_clf.predict(tfidf_matrix_test)

accuracy_score(test_df['po_ne'], preds)
```

### Step 5. 사용 버전으로 만들기

```python
lgbm_clf.predict(tfidf_vect.transform(['제 피부에 너무 잘 맞아요']))
```

```python
lgbm_clf.predict(tfidf_vect.transform(['두 통째 쓰고 있어요!']))
```

```python
def write_review():
       
    review = input('리뷰를 입력해주세요 :')

    str = [review]
    
    pred = lgbm_clf.predict(tfidf_vect.transform(str))

    if pred == [1]:
        print('---> 긍정적인 리뷰입니다 :)')

    else:
        print('---> 부정적인 리뷰입니다 :(')
```
